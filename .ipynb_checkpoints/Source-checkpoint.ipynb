```python
#importing of header files
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
#reading a csv file from url
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/bitcoin_dataset.csv')
df.head()
df.shape
```

```python
#microtask 2
#convert date to index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.info()
df.iloc[1023]
```

```python
#Microtask3 Find the corealtion
df.corr()
```

```python
#Check for null values
df.isnull().sum()
```

```python
#list of columns that have null values
null_list=['btc_total_bitcoins','btc_transaction_fees','btc_trade_volume','btc_blocks_size','btc_median_confirmation_time','btc_difficulty']
#filling the full values with for loop
for i in null_list:
    df[i]=df[i].fillna(df[i].mean())
```

```python
#cross check
df.isnull().sum()
```

```python
#splitting into 2 groups
X=df.drop(["btc_market_price"],axis=1)
Y=df["btc_market_price"]
#train ,test split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25)
```

```python
#predciting the model
linear=LinearRegression()
#to train and the model 
linear.fit(x_train,y_train)
prediction=linear.predict(x_test)
print(prediction)
```

```python
#score
score=mean_squared_error(y_test,prediction)
print(score)
```
